```python
import matplotlib.pyplot as plt

selected_labels = [
    'airplane', 'ant', 'armchair', 'bed', 'bench', 'bicycle', 'bookshelf',
    'bus', 'cabinet', 'car_sedan', 'cell_phone', 'chair', 'couch', 'door',
    'eyeglasses', 'face', 'fish', 'floor_lamp', 'flower_with_stem',
    'flying_bird', 'guitar', 'hand', 'head', 'helicopter', 'horse', 'house',
    'keyboard', 'knife', 'laptop', 'motorbike', 'piano', 'potted_plant',
    'race_car', 'satellite', 'ship', 'spider', 'spoon', 'standing_bird',
    'sword', 'table', 'tablelamp', 'telephone', 'train', 'tree', 'truck',
    'vase', 'violin', 'wheel'
]
labels_count = [
    314, 57, 333, 169, 53, 102, 202, 128, 208, 51, 170, 632, 131, 98, 92, 98,
    204, 108, 74, 94, 185, 93, 189, 162, 96, 104, 142, 152, 188, 172, 137, 169,
    60, 53, 106, 60, 78, 83, 245, 601, 90, 61, 175, 224, 52, 78, 59, 106
]
```

```python
import numpy as np

x = np.arange(len(selected_labels))
bar_height = 0.5

plt.barh(x, labels_count, tick_label=selected_labels, height=bar_height)
plt.title('Number of each label in PART-SHREC\'14')
plt.xlabel('label')
plt.ylabel('number of label')
for y, x in enumerate(labels_count):
    plt.text(x + 10, y - bar_height, '%s' % x, ha='center', va='bottom', size=5)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.show()
```
